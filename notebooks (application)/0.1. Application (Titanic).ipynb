## --- 0. Libraries Importing ---

```python
# ignore warnings
import warnings
warnings.filterwarnings("ignore")

# dataframe manipulation
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', None)

# datetime
from datetime import datetime

# set notebook width to 100%
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

# Path
import os
from pathlib import Path
os.chdir(Path(os.getcwd()).parent)

# libraries
from libraries.Visualization import (plot_distribution,
                                     plot_performances,
                                     plot_classes_performances)

from libraries.Algorithms import (DecisionTree,
                                  RandomForest)

# text coloring
from termcolor import colored

# evaluation
from sklearn.metrics import (recall_score)
```

## --- 0.1. Parameters ---

```python
uc_name = 'titanic'
target_column = 'Survived'

num_runs = 5
number_of_trees_values = [6, 7, 8, 9, 10]
handle_outliers_values = [False, True]
```

# 1. Data Loading 

```python
(X_train, X_test, y_train, y_test) = (
                                        (
                                            pd
                                            .read_csv(f'data/prepared_data/{uc_name}/X_train.csv',
                                                      low_memory=False)
                                        ),
                                        
                                        (
                                            pd
                                            .read_csv(f'data/prepared_data/{uc_name}/X_test.csv',
                                                      low_memory=False)
                                        ),
    
                                        (
                                            pd
                                            .read_csv(f'data/prepared_data/{uc_name}/y_train.csv',
                                                      low_memory=False)
                                            .loc[:, target_column]
                                        ),
                                        
                                        (
                                            pd
                                            .read_csv(f'data/prepared_data/{uc_name}/y_test.csv',
                                                      low_memory=False)
                                            .loc[:, target_column]
                                        )
                                     )

plot_distribution(df=pd.DataFrame(y_train),
                  column=target_column)
print(f"Shape -> X_train: {X_train.shape}, y_train: {y_train.shape}.")

plot_distribution(df=pd.DataFrame(y_test),
                  column=target_column)
print(f"Shape -> X_test: {X_test.shape}, y_test: {y_test.shape}.")
```

# 2. Model building 

## --- 2.1. RAFO (UC1) ---

```python
time_measurements = {}

evaluations_train_class_1 = {}
evaluations_test_class_1 = {}

evaluations_train_class_0 = {}
evaluations_test_class_0 = {}

for i in range(num_runs):
    print(f'''\t\t\t\t\t\t\t\t{colored(f'<<<<<<<< RUN {i} >>>>>>>>', 'blue', attrs=['bold'])}''')
    ########################## 0. TIME ##########################
    df_time = (pd
               .DataFrame(index=number_of_trees_values,
                          columns=handle_outliers_values)
              )
    
    df_time.index.name = 'Number Of Trees'
    df_time.columns.name = 'Handle Outliers'
    
    ########################## 1. CLASS 1 ##########################
    df_evaluation_train_class_1 = (pd
                                   .DataFrame(index=number_of_trees_values,
                                              columns=handle_outliers_values)                           
                                  )
    
    df_evaluation_train_class_1.index.name = 'Number Of Trees'
    df_evaluation_train_class_1.columns.name = 'Handle Outliers'
    
    df_evaluation_test_class_1 = (pd
                                  .DataFrame(index=number_of_trees_values,
                                             columns=handle_outliers_values)
                                 )
    
    df_evaluation_test_class_1.index.name = 'Number Of Trees'
    df_evaluation_test_class_1.columns.name = 'Handle Outliers'
    
    ########################## 2. CLASS 0 ##########################
    df_evaluation_train_class_0 = (pd
                                   .DataFrame(index=number_of_trees_values,
                                              columns=handle_outliers_values)                           
                                  )
    
    df_evaluation_train_class_0.index.name = 'Number Of Trees'
    df_evaluation_train_class_0.columns.name = 'Handle Outliers'
    
    df_evaluation_test_class_0 = (pd
                                  .DataFrame(index=number_of_trees_values,
                                             columns=handle_outliers_values)
                                 )
    
    df_evaluation_test_class_0.index.name = 'Number Of Trees'
    df_evaluation_test_class_0.columns.name = 'Handle Outliers'

    for number_of_trees in number_of_trees_values:
        for handle_outliers in handle_outliers_values:
            time_started = datetime.now()
            
            rafo = RandomForest(number_of_trees=number_of_trees,
                                max_depth=5,
                                mtry=5,
                                handle_outliers=handle_outliers)
            print(rafo)
        
            rafo.fit(X_train, y_train)
        
            train_predictions = rafo.predict(X_train)
            test_predictions = rafo.predict(X_test)
            
            time_finished = datetime.now()
        
            df_evaluation_train_class_1.loc[number_of_trees, handle_outliers] = recall_score(y_true=y_train.values,
                                                                                             y_pred=train_predictions,
                                                                                             pos_label=1)
            
            df_evaluation_train_class_0.loc[number_of_trees, handle_outliers] = recall_score(y_true=y_train.values,
                                                                                             y_pred=train_predictions,
                                                                                             pos_label=0)
            
            df_evaluation_test_class_1.loc[number_of_trees, handle_outliers] = recall_score(y_true=y_test.values,
                                                                                            y_pred=test_predictions,
                                                                                            pos_label=1)
            
            df_evaluation_test_class_0.loc[number_of_trees, handle_outliers] = recall_score(y_true=y_test.values,
                                                                                            y_pred=test_predictions,
                                                                                            pos_label=0)
            
            df_time.loc[number_of_trees, handle_outliers] = ((time_finished - time_started).seconds / 60)
        
    print(f'''\t\t\t\t\t\t\t\t{colored(f'*** RECALL -> TRAIN SET ***', 'blue', attrs=['bold'])}''')
    plot_classes_performances(df_c0=df_evaluation_train_class_0,
                              df_c1=df_evaluation_train_class_1,
                              df_t=df_time,
                              x_label='Handle Outliers',
                              y_label='Number Of Trees',
                              title_0='Class 0',
                              title_1='Class 1',
                              title_t='Time Measurements (In Minutes)')
    
    print(f'''\t\t\t\t\t\t\t\t{colored(f'*** RECALL -> TEST SET ***', 'blue', attrs=['bold'])}''')
    plot_classes_performances(df_c0=df_evaluation_test_class_0,
                              df_c1=df_evaluation_test_class_1,
                              df_t=df_time,
                              x_label='Handle Outliers',
                              y_label='Number Of Trees',
                              title_0='Class 0',
                              title_1='Class 1',
                              title_t='Time Measurements (In Minutes)')
    
    evaluations_train_class_1[f'num_run: {i}'] = df_evaluation_train_class_1
    evaluations_test_class_1[f'num_run: {i}'] = df_evaluation_test_class_1
    
    evaluations_train_class_0[f'num_run: {i}'] = df_evaluation_train_class_0
    evaluations_test_class_0[f'num_run: {i}'] = df_evaluation_test_class_0
    
    time_measurements[f'num_run: {i}'] = df_time
    
evaluations_train_class_1_avg = (pd
                                 .DataFrame(index=number_of_trees_values,
                                            columns=handle_outliers_values,
                                            data=0)
                                )

evaluations_test_class_1_avg = (pd
                                .DataFrame(index=number_of_trees_values,
                                           columns=handle_outliers_values,
                                           data=0)
                               )

evaluations_train_class_0_avg = (pd
                                 .DataFrame(index=number_of_trees_values,
                                            columns=handle_outliers_values,
                                            data=0)
                                )

evaluations_test_class_0_avg = (pd
                                .DataFrame(index=number_of_trees_values,
                                           columns=handle_outliers_values,
                                           data=0)
                               )

time_measurements_avg = (pd
                         .DataFrame(index=number_of_trees_values,
                                    columns=handle_outliers_values,
                                    data=0)
                        )

for i in range(num_runs):
    evaluations_train_class_1_avg = (evaluations_train_class_1_avg + evaluations_train_class_1[f'num_run: {i}'])
    evaluations_test_class_1_avg = (evaluations_test_class_1_avg + evaluations_test_class_1[f'num_run: {i}'])
    evaluations_train_class_0_avg = (evaluations_train_class_0_avg + evaluations_train_class_0[f'num_run: {i}'])
    evaluations_test_class_0_avg = (evaluations_test_class_0_avg + evaluations_test_class_0[f'num_run: {i}'])
    time_measurements_avg = (time_measurements_avg + time_measurements[f'num_run: {i}'])
    
evaluations_train_class_1_avg = (evaluations_train_class_1_avg / num_runs)
evaluations_test_class_1_avg = (evaluations_test_class_1_avg / num_runs)
evaluations_train_class_0_avg = (evaluations_train_class_0_avg / num_runs)
evaluations_test_class_0_avg = (evaluations_test_class_0_avg / num_runs)
time_measurements_avg = (time_measurements_avg / num_runs)

print(f'''\t\t\t\t\t\t\t\t{colored(f'*** AVERAGE TRAIN RECALL THROUGH {num_runs} RUNS ***', 'blue', attrs=['bold'])}''')
plot_classes_performances(df_c0=evaluations_train_class_0_avg,
                          df_c1=evaluations_train_class_1_avg,
                          df_t=time_measurements_avg,
                          x_label='Handle Outliers',
                          y_label='Number Of Trees',
                          title_0='Class 0',
                          title_1='Class 1',
                          title_t='Time Measurements (In Minutes)')

print(f'''\t\t\t\t\t\t\t\t{colored(f'*** AVERAGE TEST RECALL THROUGH {num_runs} RUNS ***', 'blue', attrs=['bold'])}''')
plot_classes_performances(df_c0=evaluations_test_class_0_avg,
                          df_c1=evaluations_test_class_1_avg,
                          df_t=time_measurements_avg,
                          x_label='Handle Outliers',
                          y_label='Number Of Trees',
                          title_0='Class 0',
                          title_1='Class 1',
                          title_t='Time Measurements (In Minutes)')
```

# 3. Impact calculation

## --- 3.1. Class 0 ---

```python
df_impact_class_0 = (evaluations_test_class_0_avg
                     .assign(Absolute_Performance_Difference=lambda x: x[True] - x[False])
                     .assign(Relative_Performance_Difference=lambda x: (x['Absolute_Performance_Difference'] / x[False])*100)
                    )

print(f"Mean Absolute Performance Difference (True - False): {round(df_impact_class_0['Absolute_Performance_Difference'].mean(), 3)}.")
print(f"Mean Relative Performance Difference (True - False): {round(df_impact_class_0['Relative_Performance_Difference'].mean(), 3)} %.")
```

## --- 3.2. Class 1 ---

```python
df_impact_class_1 = (evaluations_test_class_1_avg
                     .assign(Absolute_Performance_Difference=lambda x: x[True] - x[False])
                     .assign(Relative_Performance_Difference=lambda x: (x['Absolute_Performance_Difference'] / x[False])*100)
                    )

print(f"Mean Absolute Performance Difference (True - False): {round(df_impact_class_1['Absolute_Performance_Difference'].mean(), 3)}.")
print(f"Mean Relative Performance Difference (True - False): {round(df_impact_class_1['Relative_Performance_Difference'].mean(), 3)} %.")
```

## --- 3.3. Time Measurement ---

```python
df_implact_time = (time_measurements_avg
                   .assign(Absolute_Performance_Difference=lambda x: x[True] - x[False])
                   .assign(Relative_Performance_Difference=lambda x: (x['Absolute_Performance_Difference'] / x[False])*100)
                  )

print(f"Mean Absolute Performance Difference (True - False): {round(df_implact_time['Absolute_Performance_Difference'].mean(), 2)} minutes.")
print(f"Mean Relative Performance Difference (True - False): {round(df_implact_time['Relative_Performance_Difference'].mean(), 2)} %.")
```
